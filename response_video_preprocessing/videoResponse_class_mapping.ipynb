```python
response_mappings = {
    "Well": 0,
    "Poorly": 1,
    "${q://QID162/ChoiceDescription/4}": 0,
    "${q://QID162/ChoiceDescription/5}": 1
}

participant_response_mappings = {}

for participant in sorted(participants):
    participant_path = study_data_path + participant + "/mp4StudyVideo/"
    participant_videos = [video for video in os.listdir(participant_path) if video not in files_to_ignore]
    participant_response_mappings[participant] = {}
    for index, row in qualtrics_df.iterrows():
        if participant == qualtrics_df.at[index, "randomID"]:
            for col in responseColumns:
                response = qualtrics_df.at[index, col]
                for question_index, reaction_video in enumerate(warmupReactionVideoNumbers):
                    question_number = "q_warmup_"
                    try:
                        if "QID" + str(int(col[1:]) - 2) in reaction_video[0]:
                            participant_response_mappings[participant][reaction_video] = (question_number + str(question_index+3), response_mappings[response])
                    except Exception as e:
                        print(participant, col, reaction_video, response)
                for question_index, reaction_video in enumerate(reactionVideoNumbers):
                    question_number = "q_"
                    try:
                        if "QID" + str(int(col[1:]) - 2) in reaction_video[0]:
                            participant_response_mappings[participant][reaction_video] = (question_number + str(question_index+1), response_mappings[response])
                    except Exception as e:
                        print(participant, col, reaction_video, response)
participant_response_mappings
```

```python
for participant, responseVideos in participant_response_mappings.items():
    participant_path = study_data_path + participant + "/mp4StudyVideo/"
    participant_videos = os.listdir(participant_path)
    if participant == "3339":
        continue
    
    for video in sorted(participant_videos):
        mapping = video.split("_")[0]
        for responses, prediction in responseVideos.items():
            print(video, responses)
            if responses[0] == mapping:
                oldVideo_path = participant_path + video
                new_mainVideo_name = prediction[0] + "_main_" + str(prediction[1]) + ".mp4"
                os.rename(oldVideo_path, f"{participant_path}{new_mainVideo_name}")
            elif responses[1] == mapping:
                print(responses[1])
                oldVideo_path = participant_path + video
                new_mainVideo_name = prediction[0] + "_full_" + str(prediction[1]) + ".mp4"
                os.rename(oldVideo_path, f"{participant_path}{new_mainVideo_name}")
```
